# Funnel metadynamics pipeline

Launches FM pipeline, locally or remotely.

1. **Pipeline training**

   The main process that runs and analyzes experiments
   - Protein parameterization
   - Parameterization of ligands (in parallel for each ligand)
   - Creating replicas

2. **Replica**

   Calculation of a specific ligand (several replicas per 1 lignade for statistics). Each replica is a separate unit, the state of which is calculated independently of the others.
   - Addition of lignades to the receptor
   - Creating a box
   - Minimization of em energy
   - Balancing of states, nvt/npt
   - Split into individual images
   - Metadynamics for each frame (sequentially)
   - Analysis of all dG

3. **Pipeline analysis**

   Analysis of binding energies and dG errors


```python
from pathlib import Path
from simdel import run, pipelines
```

```python
pipeline = pipelines.FunnelSplit(
    workdir=Path("temp_FunnellSplit"),
    compress=True,
    workers=1,
    raw_prot_path=Path("examples/data/Funnel/pfkfb3.pdb"),
    raw_ligs_sdf=Path("examples/data/Funnel/pfkfb4_lig43.sdf"),
    extra_sdf_paths=[],
    # Fast
    n_replicas=1,
    emtol=10000,
    meta_time=0.004,
    npt_time=0.002,
    min_time=0.002,
    discard_time=0.001,
)
```
